```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
get_ipython().run_line_magic('matplotlib', 'inline')
```

```python
df = pd.read_csv("C:/Users/User10/PycharmProjects/BPSversion1/skin/dataset/resultData64.csv")
```

```python
df.head()
```

```python
x = range(len(df))
y = df['0']
fig = plt.figure(figsize=(9, 5))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
ax.barh(x, y)
for index, value in enumerate(y):
   plt. text(value, index, str("{:.0%}".format(value)))
ax.set_ylabel('Posibility Of Not Bankrupt')
ax.set_xlabel('Company')
ax.set_title('Posibility Bankruptcy Prediction Result')
ax.set_facecolor("lavender")

# Tweak spacing to prevent clipping of tick-labels
plt.subplots_adjust(bottom=0.15)
ax.legend(labels=['Not Bankrupt'], fontsize=8)
```

```python
df1 = pd.read_excel("C:/Users/User10/PycharmProjects/BPSversion1/skin/dataset/data64.xlsx")
```

```python
df1.head()
```

```python
# using some dummy data for this example
ind = range(len(df1))
attr = df1['Attr1']

fig = plt.figure(figsize=(15, 10))
ax = fig.add_axes([0.1, 0.1, 0.8, 0.8])
# 'bo-' means blue color, round points, solid lines
ax.plot(ind,attr, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
ax.set_ylabel('Attribute')
ax.set_xlabel('Company')
ax.set_title('Attribute Visulization Plot')
ax.set_facecolor("lavender")
ax.grid(True)

plt.xticks(ind, rotation='horizontal')

# Tweak spacing to prevent clipping of tick-labels
plt.subplots_adjust(bottom=0.15)
# zip joins x and y coordinates in pairs
for x,y in zip(ind,attr):

    label = "{:.3f}".format(y)

    plt.annotate(label, # this is the text
                 (x,y), # this is the point to label
                 textcoords="offset points", # how to position the text
                 xytext=(0,10), # distance from text to points (x,y)
                 ha='center') # horizontal alignment can be left, right or center

plt.show()


```
