```python
labels64 = df64.columns[0:64]
x64 = df64[labels64]
y64 = df64['target']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectFromModel
from sklearn.metrics import accuracy_score
```

```python
x_train64, x_test64, y_train64, y_test64 = train_test_split(x64,y64, test_size = 0.2, random_state = 0)

print("Number transactions X_train64 dataset: ", x_train64.shape)
print("Number transactions y_train64 dataset: ", y_train64.shape)
print("Number transactions X_test64 dataset: ", x_test64.shape)
print("Number transactions y_test64 dataset: ", y_test64.shape)
```

```python
from imblearn.over_sampling import SMOTE

print("Before OverSampling, counts of label '1': {}".format(sum(y_train64==1)))
print("Before OverSampling, counts of label '0': {} \n".format(sum(y_train64==0)))

sm = SMOTE(random_state=2)
x_train64_res, y_train64_res = sm.fit_sample(x_train64, y_train64.ravel())

print('After OverSampling, the shape of train_X64: {}'.format(x_train64_res.shape))
print('After OverSampling, the shape of train_y64: {} \n'.format(y_train64_res.shape))

print("After OverSampling, counts of label '1': {}".format(sum(y_train64_res==1)))
print("After OverSampling, counts of label '0': {}".format(sum(y_train64_res==0)))
```

```python
model64 = RandomForestClassifier(n_estimators= 100)
model64.fit(x_train64_res, y_train64_res)
```

```python
test_accuracy64 = model64.score(x_test64,y_test64)
print("Accuracy: %.2f%%" % (test_accuracy64 * 100.0))
```

```python
y_pred64 = model64.predict(x_test64)

accuracy64 = accuracy_score(y_test64, y_pred64)
print("Accuracy: %.2f%%" % (accuracy64 * 100.0))
```

```python
from joblib import dump, load
dump(model64,'BPSRFmodel64.joblib')
```
